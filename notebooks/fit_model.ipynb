```python
def _make_prior(data, prior):
    new_prior = {}
    for key in prior:
        new_prior[key] = prior[key]
    for key in ['m_pi', 'm_k', 'lam_chi', 'a/w']:
        new_prior[key] = data[key]
    return new_prior
```
